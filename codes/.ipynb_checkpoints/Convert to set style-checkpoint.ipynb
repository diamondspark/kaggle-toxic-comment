```python
import pandas as pd
```

```python
df_train = pd.read_csv('./../data/train.csv')
# df_test = pd.read_csv('./../data/test_labels.csv')
df_train.head()
```

```python
labels2 = list(df_train.columns.values)
label_dict = {}
for count, label in enumerate(labels2):
    label_dict[count] = label
label_dict
```

```python
df2= df_train
from collections import defaultdict
data = defaultdict(dict)
for row in df2.itertuples():
    data[row[0]]['SentimentText'] = row[2]
    data[row[0]]['Multilabel_sentiment'] = set()
    labels = row[4:]
    for i, label in enumerate(labels):
        if label == 1:
            data[row[0]]['Multilabel_sentiment'].add(label_dict[i+3])
```

```python
df3 = pd.DataFrame.from_dict(data, orient='index')
df3.to_pickle('./../data/df_train_set.pkl')
df3.head(15)
```
